```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.models as models
from torch.utils.data import DataLoader, Dataset
from PIL import Image
from google.colab import files
import io
import numpy as np

# Upload multiple images
uploaded = files.upload()

# Define image preprocessing for ViT
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# Custom Dataset for multiple images
class ImageDataset(Dataset):
    def __init__(self, uploaded_files, transform=None):
        self.images = []
        self.labels = []  # Dummy labels (0 or 1 for binary classification)
        self.filenames = []
        self.transform = transform

        for i, (filename, file_data) in enumerate(uploaded_files.items()):
            image = Image.open(io.BytesIO(file_data)).convert("RGB")
            if self.transform:
                image = self.transform(image)
            self.images.append(image)
            self.labels.append(i % 2)  # Assigning labels (modify for actual classes)
            self.filenames.append(filename)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        return self.images[idx], self.labels[idx], self.filenames[idx]

# Load dataset and DataLoader
dataset = ImageDataset(uploaded, transform=transform)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# Load Pretrained ViT Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vit_model = models.vit_b_16(weights=models.ViT_B_16_Weights.DEFAULT).to(device)

# **Feature Extraction Function Without `forward_features()`**
def extract_features(dataloader, model):
    model.eval()
    features = {}
    with torch.no_grad():
        for images, labels, filenames in dataloader:
            images = images.to(device)

            # Extract features using ViT
            outputs = model(images)  # Get model predictions (logits)

            # Extract features from the penultimate layer (before classifier)
            cls_embeddings = outputs[:, :768]  # Shape: (batch_size, 768)

            for i, filename in enumerate(filenames):
                features[filename] = cls_embeddings[i].cpu().numpy()  # Store as numpy array

    return features

# Extract features
image_features = extract_features(dataloader, vit_model)
print(f"Extracted features for {len(image_features)} images.")

# **Fine-Tuning ViT for Classification**
num_ftrs = vit_model.heads.head.in_features
vit_model.heads.head = nn.Linear(num_ftrs, 2)  # Modify for 2-class classification
vit_model = vit_model.to(device)

# Loss function & optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(vit_model.parameters(), lr=0.001)

# **Training Loop**
epochs = 5
for epoch in range(epochs):
    correct = 0
    total = 0
    running_loss = 0.0

    for images, labels, _ in dataloader:
        images, labels = images.to(device), torch.tensor(labels, dtype=torch.long).to(device)

        optimizer.zero_grad()
        outputs = vit_model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # Compute accuracy
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        running_loss += loss.item()

    accuracy = 100 * correct / total
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(dataloader):.4f}, Accuracy: {accuracy:.2f}%")

print("Feature extraction and training complete!")
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
import torchvision.models as models
from torch.utils.data import DataLoader, Dataset
from PIL import Image
from google.colab import files
import io
import numpy as np
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# ✅ 1️⃣ Upload Images
uploaded = files.upload()

# ✅ 2️⃣ Define Image Preprocessing for ViT
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# ✅ 3️⃣ Custom Dataset for Multiple Images
class ImageDataset(Dataset):
    def __init__(self, uploaded_files, transform=None):
        self.images = []
        self.labels = []  # Dummy labels (modify for actual classes)
        self.filenames = []
        self.transform = transform

        for i, (filename, file_data) in enumerate(uploaded_files.items()):
            image = Image.open(io.BytesIO(file_data)).convert("RGB")
            if self.transform:
                image = self.transform(image)
            self.images.append(image)
            self.labels.append(i % 2)  # Assigning labels (modify for actual classes)
            self.filenames.append(filename)

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        return self.images[idx], self.labels[idx], self.filenames[idx]

# ✅ 4️⃣ Load Dataset and DataLoader
dataset = ImageDataset(uploaded, transform=transform)
dataloader = DataLoader(dataset, batch_size=4, shuffle=True)

# ✅ 5️⃣ Load Pretrained ViT Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vit_model = models.vit_b_16(weights=models.ViT_B_16_Weights.DEFAULT).to(device)

# ✅ 6️⃣ Correct Feature Extraction from CLS Token
def extract_features(dataloader, model):
    model.eval()
    features = {}

    with torch.no_grad():
        for images, labels, filenames in dataloader:
            images = images.to(device)

            # Forward pass through ViT
            outputs = model(images)  # Get model output

            # Extract CLS token from final layer
            cls_embeddings = outputs[:, :768]  # Shape: (batch_size, 768)

            for i, filename in enumerate(filenames):
                features[filename] = cls_embeddings[i].cpu().numpy()

    return features

# ✅ 7️⃣ Extract Features
image_features = extract_features(dataloader, vit_model)
print(f"Extracted features for {len(image_features)} images.")

# ✅ 8️⃣ Modify ViT Model for Classification
num_ftrs = vit_model.heads.head.in_features
vit_model.heads.head = nn.Linear(num_ftrs, 2)  # 2-class classification
vit_model = vit_model.to(device)

# ✅ 9️⃣ Define Loss & Optimizer
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(vit_model.parameters(), lr=0.0001)

# ✅ 🔟 Training Loop
epochs = 5
for epoch in range(epochs):
    correct = 0
    total = 0
    running_loss = 0.0

    for images, labels, _ in dataloader:
        images, labels = images.to(device), torch.tensor(labels, dtype=torch.long).to(device)

        optimizer.zero_grad()
        outputs = vit_model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        # Compute Accuracy
        _, predicted = torch.max(outputs, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
        running_loss += loss.item()

    accuracy = 100 * correct / total
    print(f"Epoch [{epoch+1}/{epochs}], Loss: {running_loss/len(dataloader):.4f}, Accuracy: {accuracy:.2f}%")

print("🎯 Training Complete!")

# ✅ 1️⃣1️⃣ Evaluate Model Performance
def evaluate_model(dataloader, model):
    y_true = []
    y_pred = []

    model.eval()
    with torch.no_grad():
        for images, labels, _ in dataloader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            y_true.extend(labels.cpu().numpy())
            y_pred.extend(predicted.cpu().numpy())

    acc = accuracy_score(y_true, y_pred)
    print("\n🎯 Model Accuracy: {:.2f}%".format(acc * 100))
    print("\n📊 Classification Report:\n", classification_report(y_true, y_pred))

    # Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt="d", cmap="Blues", xticklabels=["Class 0", "Class 1"], yticklabels=["Class 0", "Class 1"])
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.title("Confusion Matrix")
    plt.show()

# ✅ Run Evaluation
evaluate_model(dataloader, vit_model)

# ✅ 1️⃣2️⃣ Save the Trained Model
torch.save(vit_model.state_dict(), "vit_trained_model.pth")
print("✅ Model Saved!")

# ✅ 1️⃣3️⃣ Inference on New Images
def predict_image(image_path, model):
    transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
    ])

    image = Image.open(image_path).convert("RGB")
    image = transform(image).unsqueeze(0).to(device)  # Add batch dimension

    model.eval()
    with torch.no_grad():
        output = model(image)
        _, predicted = torch.max(output, 1)

    class_names = ["Class 0", "Class 1"]
    print(f"🖼️ Predicted Class: {class_names[predicted.item()]}")

# ✅ Upload an image for inference
uploaded = files.upload()
for filename in uploaded.keys():
    predict_image(filename, vit_model)
```

```python
import pandas as pd
import torch
import torchvision.transforms as transforms
from PIL import Image
from google.colab import files
import io

# ✅ 1️⃣ Define Actual Class Labels
class_labels = {0: "Forensic Evidence", 1: "Non-Forensic"}

# ✅ 2️⃣ Load the Trained Model
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
vit_model.load_state_dict(torch.load("vit_trained_model.pth", map_location=device))
vit_model = vit_model.to(device)
vit_model.eval()

# ✅ 3️⃣ Image Preprocessing for Inference
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])
])

# ✅ 4️⃣ Predict Function with CSV Logging
def predict_and_log():
    uploaded = files.upload()
    results = []

    for filename, file_data in uploaded.items():
        # Read and preprocess image
        image = Image.open(io.BytesIO(file_data)).convert("RGB")
        image = transform(image).unsqueeze(0).to(device)

        # Perform inference
        with torch.no_grad():
            output = vit_model(image)
            _, predicted = torch.max(output, 1)
            predicted_class = class_labels[predicted.item()]

        print(f"🖼️ {filename} → Predicted Class: {predicted_class}")
        results.append({"Filename": filename, "Predicted Class": predicted_class})

    # Save results to CSV
    df = pd.DataFrame(results)
    df.to_csv("predictions.csv", index=False)
    print("✅ Predictions saved to predictions.csv!")

# ✅ Run Prediction & Save Results
predict_and_log()
```
